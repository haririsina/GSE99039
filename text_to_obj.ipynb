### imports 

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

### reading data from text into df

```python
data_path = '/Users/sina/JupyterNotebook/Hostpital/data/GSE99039_series_matrix.txt'
raw_df = pd.read_csv(data_path, sep='\t', comment='!', header=None, low_memory=False)
raw_df.shape
```

```python
raw_df.head(18)
```

## dealing with null values 

```python
print(raw_df.isnull().sum().sum())
df = raw_df.fillna(value='#')
print(df.isnull().sum().sum())
```

## transfering data from df to objects

```python
headers = [
    'sample_title',
    'disease',
    'learning_set',
    'sex',
    'genetic',
    'age_at_first_symptoms',
    'updrs_i',
    'updrs_ii',
    'updrs_iii_(on)',
    'updrs_iii_(off)',
    'updrs_iv',
    'hoehn_yahr_(on)',
    'hoehn_yahr_(off)',
    'mocca_score',
    'age_at_exam',
    'id_ref',
    'mRNA_expression',
]

# list for storing sample objects
sample_list = []

# # itter over columns
for indx, series in tqdm(df.iteritems()):
    sample_object = {}
    mRNA = []
    
    # # skipping the first column 
    if indx==0:
        continue
    
    # # itter over rows
    for i, ob in enumerate(series):
        
        if i <=15:
            # # creating sample_obj
            sample_object[headers[i]] = ob
        else:
            mRNA.append(ob)
            
    sample_object['mRNA_expression'] = mRNA
    sample_list.append(sample_object)
```

### render JSON

```python
import uuid
from IPython.display import display_javascript, display_html, display
import json

class RenderJSON(object):
    def __init__(self, json_data):
        if isinstance(json_data, dict):
            self.json_str = json.dumps(json_data)
        else:
            self.json_str = json_data
        self.uuid = str(uuid.uuid4())

    def _ipython_display_(self):
        display_html('<div id="{}" style="height: 600px; width:100%;"></div>'.format(self.uuid), raw=True)
        display_javascript("""
        require(["https://rawgit.com/caldwell/renderjson/master/renderjson.js"], function() {
        document.getElementById('%s').appendChild(renderjson(%s))
        });
        """ % (self.uuid, self.json_str), raw=True)
```
